```python
##This is where I'm starting my answers##

##Load data and plot/compare variables

filename="mytraining.csv"

RawVars=!head -1 /data/afarbin/DLClass/HiggsML/training.csv #this returns the names as an SList size 1
VarNames=str.split(RawVars[0], ",")
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv(filename, names=VarNames) #this gives a "mixed type" sort-of error that doesn't actually break anything
                                           #asks to specify dtype or set low_memory=False
```

```python
df
```

```python
#normalization
for var in VarNames[1:len(VarNames)-2]: #the weight can stay as it is
    cur=df[var]
    df[var]=(cur-cur.min())/(cur.max()-cur.min())
    #go through and normalize: (x-min)/(max-min)

df
```

```python
#well all of that seems okay

df_sig=df[df.Label=='s']
df_bkg=df[df.Label=='b']
```

```python
for var in VarNames[1:len(VarNames)-2]: #this is ugly but correct: cut out the EventID at front and Label at the end
    print var                           #and I guess we don't need the weight for this
    plt.figure()
    print(df_sig[var].max(), df_sig[var].min())
    plt.hist(df_sig[var],bins=100,histtype="step", color="red",label="background",stacked=True)
    plt.hist(df_bkg[var],bins=100,histtype="step", color="blue", label="signal",stacked=True)
    plt.legend(loc='upper right')
    plt.show()
```

```python
##Setup/train classifier

import sklearn.discriminant_analysis as DA
Fisher=DA.LinearDiscriminantAnalysis()
```

```python
#also copied from lab 4
N_Train=12500

Train_Sample=df[:N_Train]
Test_Sample=df[N_Train:]

X_Train=Train_Sample[VarNames[1:len(VarNames)-2]]
y_Train=Train_Sample["Label"]

X_Test=Test_Sample[VarNames[1:len(VarNames)-2]]
y_Test=Test_Sample["Label"]

Test_sig=Test_Sample[Test_Sample.Label=='s']
Test_bkg=Test_Sample[Test_Sample.Label=='b']

#Efficiency: take number of things that get through and sum the weights, and that is the efficiency - not
#the number of things itself
```

```python
Fisher.fit(X_Train,y_Train)
```

```python
plt.figure()
plt.hist(Fisher.decision_function(Test_sig[VarNames[1:len(VarNames)-2]]),bins=100,histtype="step", color="blue", label="signal",stacked=True)
plt.hist(Fisher.decision_function(Test_bkg[VarNames[1:len(VarNames)-2]]),bins=100,histtype="step", color="red", label="background",stacked=True)
plt.legend(loc='upper right')
plt.show()
```

```python
#doin the roc curve

from sklearn.metrics import roc_curve, auc
fpr, tpr, _ = roc_curve(y_Test, Fisher.decision_function(X_Test))

roc_auc = auc(fpr, tpr)

plt.plot(fpr,tpr,color='darkorange',label='ROC curve (area = %0.2f)' % roc_auc)
plt.legend(loc="lower right")
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')

plt.show()
#how is this data different from lab 4 data?
#what makes this data give the "ValueError: Data is not binary and pos_label is not specified" error?
#I dunno what I'm doing \o/ so let's move on to something else
```

```python
#observation significance calculation thingy
#there was something about changing the formula

import numpy as np

def PlotSignificance(N_S,N_B, N_S_min=1):
    plt.figure()
    eff_sig,bins_sig,p_sig=plt.hist(Fisher.decision_function(Test_sig[VarNames[1:len(VarNames)-2]]),bins=100,histtype="step", color="blue", label="signal",cumulative=-1,stacked=True,normed=True)
    eff_bkg,bins_bkg,p_bkg=plt.hist(Fisher.decision_function(Test_bkg[VarNames[1:len(VarNames)-2]]),bins=100,histtype="step", color="red", label="background",cumulative=-1,stacked=True,normed=True)
    plt.legend(loc='upper right')
    plt.show()
    
    good_bins = np.where(eff_sig*N_S>=N_S_min)

    print len(good_bins[0])
    if len(good_bins[0])<1:
        print "Insufficient Signal."
        return 0,0,0
    
    significance=(N_S*eff_sig)/np.sqrt((N_B*eff_bkg)+(N_S*eff_sig))
    #ravioli ravioli give me the formuoli
    #note to self: fix this when I manage to locate the proper formula
    

    plt.figure()
    plt.plot(bins_sig[:-1],significance)
    
    max_sign=np.max(significance[good_bins])
    max_signI=np.argmax(significance[good_bins])
    
    plt.show()
    print "Max significance at ", bins_sig[max_signI], " of", max_sign
    return bins_sig[max_signI],max_sign, max_signI
    
PlotSignificance(1000000,1e11)
```
